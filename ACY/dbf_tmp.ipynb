```python
import re
import pandas as pd

with open(f'c:\History.txt', 'r', encoding='utf-8') as f:
    text = f.read()
text = text[1:]
t1 = text.split('================================')
t1.pop()

arr = []
p = re.compile(r'\[.*\]')

for w in t1:
    ls = w.split('\n')
    if '[Yandex > English to Russian]' in ls:
        index = ls.index('[Yandex > English to Russian]')
        en = ls[index-2]
        ru = ls[index+1]
        tr = ''

        if len(ls) > (index+3):
            st = ls[index+3]
            ls1 = p.findall(st)
            if len(ls1) > 0:
                tr = ls1[0]
            if len(en.split(' ')) < 3:
                arr.append([en, tr, ru, '\n'.join(ls[index+3:-3])])

df = pd.DataFrame(arr, columns=['En', 'Trans', 'Rus', 'Full'])
df.fillna('', inplace=True)


with open(f'd:\GoogleDrive\History.txt ', mode='w', encoding='utf-8') as f:
    for ls in df.values:
        f.write('###\n')
        f.write(ls[0] + '\n')
        f.write(ls[1] + '\n')
        f.write(ls[2] + '\n')
        f.write('####\n')
        f.write(ls[3] + '\n')
    f.write('###\n')
```
