```python
def get_mailbox(mailbox: IMAP4, mailbox_name: str) -> IMAP4:
    """Get the mailbox object for the given mailbox name."""
    status, data = mailbox.select(mailbox_name)
    if status != 'OK':
        raise ValueError(f'Could not select mailbox {mailbox_name}')
    return mailbox
```

```python
def get_unread_messages(mailbox: IMAP4) -> list:
    """Return a list of unread messages in the given mailbox."""
    status, data = mailbox.search(None, 'UNSEEN')
    if status != 'OK':
        raise ValueError('Could not search mailbox')
    return data[0].split()
```

```python
def get_message(mailbox: IMAP4, message_id: bytes) -> str:
    """Return the body of the given message."""
    status, data = mailbox.fetch(message_id, '(RFC822)')
    if status != 'OK':
        raise ValueError('Could not fetch message')
    return data[0][1].decode('utf-8')
```

```python
def get_messages(mailbox: IMAP4) -> list:
    """Return a list of unread messages in the given mailbox."""
    unread = get_unread_messages(mailbox)
    return [get_message(mailbox, message_id) for message_id in unread]
```

```python
def delete_message(mailbox: IMAP4, message_id: bytes) -> None:
    """Delete the given message from the given mailbox."""
    status, data = mailbox.store(message_id, '+FLAGS', '\\Deleted')
    if status != 'OK':
        raise ValueError('Could not delete message')
```

```python
def delete_messages(mailbox: IMAP4) -> None:
    """Delete all messages from the given mailbox."""
    for message_id in get_unread_messages(mailbox):
        delete_message(mailbox, message_id)
```

```python
def close_mailbox(mailbox: IMAP4) -> None:
    """Close the given mailbox."""
    mailbox.close()
```

```python
def logout(mailbox: IMAP4) -> None:
    """Log out of the given mailbox."""
    mailbox.logout()
```
