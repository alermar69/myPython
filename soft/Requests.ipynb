```python
import requests
import json
import pandas as pd
```

```python
def api_get_token(username, password):
    url = 'http://localhost:8000/ru/api-token-auth/'
    PARAMS = {
        "username": username,
        "password": password
    }
    r = requests.post(url = url, data = PARAMS)
    data = r.json()
    return data['token']


def api_get(url, token):
    headers = {
        'Accept': 'application/json',
        'Authorization': f'Token {token}'
    }
    r = requests.get(url = url, headers=headers)
    data = r.json()
    return data
```
